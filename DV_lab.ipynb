```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import parallel_coordinates

# Load dataset
df = pd.read_csv("/content/clean_dataset .csv")

# Select required features
features = ["Age", "Debt", "YearsEmployed"]

# Drop rows with missing values in required columns
df = df.dropna(subset=features + ["PriorDefault"])

# Convert hue column to categorical (important!)
df["PriorDefault"] = df["PriorDefault"].astype(str)

# Create pairplot
sns.pairplot(
    df,
    vars=features,
    hue="PriorDefault",
    diag_kind="hist",
    height=2.5,

)

# Add title
plt.suptitle(
    "Scatterplot Matrix with Histogram on Diagonal",
    y=1.02
)

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import parallel_coordinates

# Load dataset
df = pd.read_csv("/content/clean_dataset .csv")

# Select required features
features = ["Age", "Industry", "YearsEmployed"]

# Drop rows with missing values in required columns
df = df.dropna(subset=features + ["PriorDefault"])

# Convert hue column to categorical (important!)
df["PriorDefault"] = df["PriorDefault"].astype(str)

# Create pairplot
sns.pairplot(
    df,
    vars=features,
    hue="PriorDefault",
    diag_kind="hist",
    height=2.5,

)

# Add title
plt.suptitle(
    "Scatterplot Matrix with Histogram on Diagonal",
    y=1.02
)

plt.show()
```

```python
import plotly.express as px
from sklearn.preprocessing import StandardScaler

features = ['Gender', 'Age', 'Debt', 'Married', 'BankCustomer', 'YearsEmployed', 'PriorDefault','Employed', 'CreditScore', 'DriversLicense', 'Income','Approved']

scaler = StandardScaler()
df_scaled = df.copy()
df_scaled[features] = scaler.fit_transform(df[features])
fig = px.parallel_coordinates(
    df_scaled,
    dimensions=features,
    color="PriorDefault",
    title="Normalized parallel coordinates plot"
)

fig.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/clean_dataset .csv")


binary_cols = [
    'Gender', 'Married', 'BankCustomer',
    'PriorDefault', 'Employed', 'DriversLicense', 'Approved'
]

for col in binary_cols:
    df[col] = df[col].map({'No': 0, 'Yes': 1})


features = [
    'Gender', 'Age', 'Debt', 'Married', 'BankCustomer',
    'YearsEmployed', 'PriorDefault', 'Employed',
    'CreditScore', 'DriversLicense', 'Income', 'Approved'
]


df[features].plot(kind="line", figsize=(14, 6))

plt.title("Multivariate Line Graph (All Features)")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend(loc="upper right", bbox_to_anchor=(1.15, 1))
plt.tight_layout()
plt.show()
```

```python
 import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("/content/clean_dataset .csv")


pd.crosstab(df['Gender'], df['Approved']).plot(
    kind="bar",
    stacked=True,
    figsize=(6, 4)
)


plt.title("Stacked Bar Chart: Gender vs Approved")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Approved")
plt.tight_layout()
plt.show()
```
