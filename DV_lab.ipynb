```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import parallel_coordinates

# Load dataset
df = pd.read_csv("/content/clean_dataset .csv")

# Select required features
features = ["Age", "Debt", "YearsEmployed"]

# Drop rows with missing values in required columns
df = df.dropna(subset=features + ["PriorDefault"])

# Convert hue column to categorical (important!)
df["PriorDefault"] = df["PriorDefault"].astype(str)

# Create pairplot
sns.pairplot(
    df,
    vars=features,
    hue="PriorDefault",
    diag_kind="hist",
    height=2.5,

)

# Add title
plt.suptitle(
    "Scatterplot Matrix with Histogram on Diagonal",
    y=1.02
)

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import parallel_coordinates

# Load dataset
df = pd.read_csv("/content/clean_dataset .csv")

# Select required features
features = ["Age", "Debt", "YearsEmployed"]

# Drop rows with missing values in required columns
df = df.dropna(subset=features + ["PriorDefault"])

# Convert hue column to categorical (important!)
df["PriorDefault"] = df["PriorDefault"].astype(str)

# Create pairplot
sns.pairplot(
    df,
    vars=features,
    hue="PriorDefault",
    diag_kind="kde",
    height=2.5,

)

# Add title
plt.suptitle(
    "Scatterplot Matrix with Histogram on Diagonal",
    y=1.02
)

plt.show()
```
